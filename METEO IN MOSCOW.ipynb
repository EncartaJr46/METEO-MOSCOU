```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.animation as animation
from IPython.display import HTML
```

```python
df = pd.DataFrame.from_dict(
    {
        "Jan": -6.7,
        "Feb": -6.1,
        "Mar": -0.9,
        "Apr": +6.9,
        "May": +13.3,
        "Jun": +17.2,
        "Jul": +20.0,
        "Aug": +17.9,
        "Sep": +11.8,
        "Oct": +5.9,
        "Nov": -0.1,
        "Dec": -4.3
    },
    orient="index",
    columns=["temperature_C"]
)

df
```

```python
fig, ax = plt.subplots(
    figsize=(9,6),
    facecolor="white"
)
fig.suptitle("Monthly average temperature in MOSCOW, RUSSIAN FEDERATION ")

ax.bar(
    x=df.index,
    height=df["temperature_C"]
)
ax.set_ylabel("°C")

plt.tight_layout()

plt.savefig(
    fname="plot1.png",
    format="png",
    dpi=125
);
```

```python
fig, ax = plt.subplots(
    figsize=(9,6),
    facecolor="white"
)
fig.suptitle("Monthly average temperature in Moscow, RUSSIAN FEDERATION measured in °C")

ax.bar(
    x=df.index,
    height=df["temperature_C"]
)

# Remove spines
for position in ["left", "top", "right"]:
    ax.spines[position].set_visible(False)

# Remove `y` axis ticks and tick labels
ax.tick_params(
    axis="y",
    which="major",
    left=False,
    labelleft=False
)

# Add bar data labels
for x_val, y_val in enumerate(df["temperature_C"]):
    ax.text(
        x_val, y_val,
        f"{y_val}",
        ha="center",
        va="bottom"
    )

plt.tight_layout()

plt.savefig(
    fname="plot2.png",
    format="png",
    dpi=125
);
```

```python
text_colour = "#5e5c64"  # dark 2
bar_colour = "#deddda"  # light 3
plt.rcParams["font.family"] = "Roboto"  # update font for the whole plot
col_min = "#ed333b"  # red 2
col_max = "#57e389"  # green 2

# Create a list of colours for every bar
all_colours = []
for val in df["temperature_C"]:
    if val == max(df["temperature_C"]):
        col = col_max
    elif val == min(df["temperature_C"]):
        col = col_min
    else:
        col = bar_colour
        
        

    all_colours.append(col)

fig, ax = plt.subplots(
    figsize=(9,6),
    facecolor="white"
)
fig.suptitle(
    "Monthly average temperature in MOSCOW, RUSSIAN FEDERATION measured in °C",
    fontsize=20,
    color=text_colour
    )

ax.bar(
    x=df.index,
    height=df["temperature_C"],
    color=all_colours
)

# Remove spines
for position in ["left", "top", "right"]:
    ax.spines[position].set_visible(False)

# Remove `y` axis ticks and tick labels
ax.tick_params(
    axis="y",
    which="major",
    left=False,
    labelleft=False
)

# Add bar data labels
for x_val, y_val in enumerate(df["temperature_C"]):
    if y_val == max(df["temperature_C"]):
        col = col_max
        txt = f"max {y_val}"
    elif y_val == min(df["temperature_C"]):
        col = col_min
        txt = f"min {y_val}"
    else:
        col = bar_colour
        txt = f"{y_val}"

    ax.text(
        x_val, y_val,
        txt,
        ha="center",
        va="bottom",
        fontsize=13,
        color=col
    )

# Update appearance of the `x` axis tick labels
ax.tick_params(
    axis="x",
    labelsize=14,
    labelcolor=text_colour
)

plt.tight_layout()

plt.savefig(
    fname="plot5.png",
    format="png",
    dpi=125
);
```
